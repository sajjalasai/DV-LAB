```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load dataset
# Replace 'Sheet1' with your actual sheet name if different
df = pd.read_excel("final dataset Book1.xlsx", sheet_name="Sheet1")

# Example: assuming your dataset has 'winnername' and 'losername' columns
# Replace these with your actual column names
G = nx.from_pandas_edgelist(df, source="winnername", target="losername")

# Draw the graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, k=0.5, iterations=50)  # Layout for visualization
nx.draw(G, pos, with_labels=True,
        node_color="skyblue", edge_color="gray",
        node_size=1500, font_size=10, font_weight="bold")

plt.title("Graph and Network Visualization", fontsize=14)
plt.show()

# --- Graph Analysis ---
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
print("Degree Centrality:", nx.degree_centrality(G))
print("Clustering Coefficient:", nx.average_clustering(G))
```
