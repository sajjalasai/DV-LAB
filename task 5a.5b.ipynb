```python
import pandas as pd

df = pd.read_csv('/content/Book1.csv')
```

```python
wins_by_winner = df.groupby('winnername').size().reset_index(name='wins')
display(wins_by_winner.head())
```

```python
import plotly.express as px

fig = px.treemap(
    data_frame=wins_by_winner,
    path=['winnername'],
    values='wins',
    title='Treemap of Wins by Winner'
)
fig.show()
```

```python
sunburst_data = df.groupby('winnername').size().reset_index(name='count')
display(sunburst_data.head())
```

```python
import plotly.express as px

fig = px.sunburst(
    data_frame=sunburst_data,
    path=['winnername'],
    values='count',
    title='Sunburst Chart of Wins by Winner'
)
fig.show()
```
